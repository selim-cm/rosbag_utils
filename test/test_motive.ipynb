## Header

The only useful value in header is the start time.

```python
# Start time is understood as a column in pandas
df_head.columns
```

```python
index_capture_start_time = df_head.columns.get_loc('Capture Start Time')
time = df_head.columns.to_list()[index_capture_start_time + 1]
time
```

```python
import datetime
time_datetime = datetime.datetime.strptime(time, '%Y-%m-%d %H.%M.%S.%f %p')
time_datetime
```

```python
epoch = time_datetime.timestamp()
epoch
```

## Data

The following modifications can be made to the data:
- Remove NaN columns
- Remove duplicate columns
- Replace time with epoch
- Rename columns to have one rotation and one translation column per rigid body

```python
# Remove NaN columns
df_no_head = df_no_head.dropna(axis=1, how='all')
df_no_head
```

```python
# Remove first column as it is a duplicate of the index
df_no_head = df_no_head.iloc[:, 1:]
df_no_head
```

```python
# Replace time column with epoch
df_no_head['Time (Seconds)'] += [epoch for i in range(df_no_head.shape[0])]
df_no_head
```

```python
# Check epoch values are correct
[df_no_head['Time (Seconds)'][i] for i in range(df_no_head.shape[0])]
```

```python
# Rename columns
nb_rigid_bodies = int((df_no_head.shape[1] - 1) / 7)
nb_rigid_bodies
```

```python
list_of_columns_name = ['Epoch']
for i in range(1, nb_rigid_bodies+1):
    list_of_columns_name.append('x'+str(i))
    list_of_columns_name.append('y'+str(i))
    list_of_columns_name.append('z'+str(i))
    list_of_columns_name.append('qx'+str(i))
    list_of_columns_name.append('qy'+str(i))
    list_of_columns_name.append('qz'+str(i))
    list_of_columns_name.append('qw'+str(i))
df_no_head.columns = list_of_columns_name
df_no_head
```
